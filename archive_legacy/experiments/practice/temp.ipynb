```python
import time
import torch
tic = time.time()
a = torch.rand(1000)
b = torch.rand(1000)
c = a @ b
print(c)
toc = time.time()
print(f"1:{1000*(toc-tic)} ms")


tic = time.time()
for i in range(1000):
    c += a[i]*b[i]
toc = time.time()
print(f"2:{1000*(toc-tic)} ms")
```

```python
import time
import torch

# 热身（避免第一次GPU/CPU初始化开销）
_ = torch.rand(10) @ torch.rand(10)

# 测试1：向量化
tic = time.time()
a = torch.rand(1000)
b = torch.rand(1000)
c = a @ b
toc = time.time()
print(f"1. 向量化（含创建张量）: {1000*(toc-tic):.4f} ms")

# 测试2：仅计算向量化点积（不包括创建张量）
tic = time.time()
c = a @ b
toc = time.time()
print(f"2. 纯向量化计算: {1000*(toc-tic):.4f} ms")

# 测试3：循环
tic = time.time()
c_loop = torch.tensor(0.0)
for i in range(1000):
    c_loop += a[i] * b[i]
toc = time.time()
print(f"3. 循环计算: {1000*(toc-tic):.4f} ms")
```
